# Ridge regression

Ayuda a estabilizar un modelo con errores un poco más grandes

## Paquetes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import Ridge
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, root_mean_squared_error
```

## conjunto de datos

```python
np.random.seed(42)
X = np.random.rand(200,1)*10
y = 4*X.squeeze() + np.random.randn(200)*3

df = pd.DataFrame({"X": X.squeeze(),"y": y})
```

```python
plt.scatter(X, y)
```

## Estandarización de datos

Toca estandarizar los datos para que ridge funcione

```python
ridge_model = Pipeline([
    ("scaler", StandardScaler()),
    ("ridge", Ridge(alpha=1.0))
])
```

## Entrenamiento 

```python
X_train, X_test , y_train, y_test = train_test_split(df[["X"]], df[["y"]], test_size = 0.2, random_state=42)        
```

## Modelo

```python
ridge_model.fit(X_train, y_train)
y_pred = ridge_model.predict(X_test)
```

```python
rmse = root_mean_squared_error(y_test, y_pred)
mae_ridge = mean_absolute_error(y_test, y_pred)

rmse, mae_ridge
```

## Conclusiones

Ridge no es un método para bajar error más bien es usado para vajar varianzas del modelo, sin embargo es útil en casos extraños donde el modelo tiene bastante ruido.
