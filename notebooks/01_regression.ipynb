# Regresión lineal

## 1. Objetivo del análisis

Modelar la relación entre una variable continua $X$ y una variable objetivo $y$,
utilizando regresión lineal como modelo base interpretable.

## 2. Exploración del dataset

Data Set artificial con el fin de acelerar el proceso.

```python
import numpy as np # paquete para manejo de arreglos
import pandas as pd # paquete para manejo de datos
import matplotlib.pyplot as plt # paquete para visualización

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error, mean_absolute_error

np.random.seed(42)

X = np.random.rand(200, 1)*10
y = 4*X.squeeze() + 7 + np.random.randn(200)*3

df = pd.DataFrame({'X': X.squeeze(), 'y': y})
df.head()   
```

```python
plt.scatter(X, y)
```

división entre datos de prueba y de entrenamiento, los datos de prueba son el 20 % mientras que los de entrenamiento son el 80 %

```python
X_train, X_test, y_train, y_test = train_test_split(df[["X"]], df[["y"]], test_size = 0.2, random_state=42)
```

```python
model = LinearRegression() # modelo elegido 
model.fit(X_train, y_train) # entrenamiento al que se ajusta
```

```python
y_pred = model.predict(X_test)

rmse = root_mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"ERROR CUADRADO: {rmse}")
print(f"ERROR ABSOLUTO: {mae}")
```
